```python
import pandas as pd

# Loading Scraped Data
df = pd.read_csv("/content/tweets-data/SENTIMEN(Agustus).csv")

# Filtering tweets that contain 'DPR' or 'wakil rakyat'
df = df[df["full_text"].str.contains("DPR|wakil rakyat", case=False, na=False)]

print(f"Jumlah tweet setelah filter: {len(df)}")

# Save the Filtered Results
df.to_csv("tweets_filtered.csv", index=False, encoding="utf-8")
df.head(15)
```

Data Cleaning

```python
import re

def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text).lower()
    text = re.sub(r"http\S+", " ", text)       # Remove URL
    text = re.sub(r"@\w+", " ", text)          # Remove mention
    text = re.sub(r"#\w+", " ", text)          # Remove hashtag
    text = re.sub(r"[^0-9a-zA-Zà-ÿ’' ]+", " ", text) # Remove Symbols/Emojis
    text = re.sub(r"\s+", " ", text).strip()   # Fix the spacing
    return text

df["text_clean"] = df["full_text"].apply(clean_text)

df_clean = df[[
    "created_at", "full_text", "text_clean",
    "favorite_count", "retweet_count", "reply_count", "quote_count"
]]

df_clean.to_csv("tweets_clean.csv", index=False, encoding="utf-8")
df_clean.head(15)
```

Data Classification

```python
from langchain_community.llms import Replicate
import os
from google.colab import userdata

# Set the API token
api_token = userdata.get('REPLICATE_API_TOKEN')
os.environ["REPLICATE_API_TOKEN"] = api_token

# Model setup
model = "ibm-granite/granite-3.3-8b-instruct"

output = Replicate(
    model=model,
    replicate_api_token=api_token,
)

# Refine the prompt
sample_tweets = df_clean["text_clean"].head(15).tolist()
tweets_text = "\n".join([f"Tweet {i+1}: {t}" for i, t in enumerate(sample_tweets)])

prompt_classify = f"""
Classify these tweet as positive, negative, neutral, or mixed:

{tweets_text}
"""

# Invoke the model with the example prompt
response_classify = output.invoke(prompt_classify)

# Print the response
print("Granite Classification:\n")
print(response_classify)
```

```python
# Define refined prompt
prompt_classify_refined = f"""
Analyze each tweet about DPR (Indonesian parliament) with the following rules:

1. Determine the overall sentiment: "Positive", "Negative", "Neutral", or "Mixed".
2. If "Mixed", separate clearly which parts are positive and which are negative.
3. Identify the main topic of the tweet: "Salary", "Performance", "Ethics", "Transparency", or "Other".
4. Return the result strictly in JSON list format, one object per tweet, with the following structure:
{{
  "tweet": "...",
  "sentiment": "...",
  "positive_aspects": ["..."],
  "negative_aspects": ["..."],
  "topic": "..."
}}

Tweets to classify:
{tweets_text}
"""

# Invoke the model with the refined prompt
response_refined = output.invoke(prompt_classify_refined)

# Print the response
print("Granite Model Refined Response:\n")
print(response_refined)
```

```python
#Define the prompt
prompt_classify_next = f"""
For each tweet about DPR salary, classify into the following:

1. sentiment: "Positive", "Negative", "Neutral", or "Mixed"
2. sentiment_intensity: "Weak", "Moderate", or "Strong"
3. topic: "Salary", "Performance", "Ethics", "Transparency", or "Other"
4. target: "DPR Institution", "Individual DPR Member", "Government/Policy", or "Other"

Return the result strictly in JSON list format:
{{
  "tweet": "...",
  "sentiment": "...",
  "sentiment_intensity": "...",
  "topic": "...",
  "target": "..."
}}

Tweets to classify:
{tweets_text}
"""

# Invoke the model with the refined prompt
response_next = output.invoke(prompt_classify_next)

# Print the response
print(response_next)
```

```python
prompt_classify_argument = f"""
For each tweet about DPR salary, identify the main argument type used by the author.

Categories:
- "Economic" → about salary, tax, minimum wage, state budget
- "Ethical/Moral" → about justice, fairness, ethics, morality
- "Political" → about DPR performance, policies, parties, or power
- "Emotional" → expression of anger, sarcasm, or frustration without reasoning
- "Other"

Return the result strictly in JSON list format:
{{
  "tweet": "...",
  "argument_type": "..."
}}

Tweets to classify:
{tweets_text}
"""

# Invoke the model with the refined prompt
response_argument = output.invoke(prompt_classify_argument)

# Print the response
print(response_argument)
```
